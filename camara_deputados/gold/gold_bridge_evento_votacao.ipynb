```python
from pyspark.sql import functions as F

df_votacoes = spark.table(
    "workspace.gold.fato_votacoes"
)

df_eventos = spark.table(
    "workspace.gold.fato_eventos"
)
```

```python
df_bridge = (

    df_votacoes

    .filter(
        F.col("id_evento").isNotNull()
    )

    .select(
        F.col("id_evento"),
        F.col("sk_votacao"),
        F.col("dt_votacao"),
        F.col("id_orgao")
    )

    .dropDuplicates()

)
```

```python
display(df_bridge)
```

```python
(
    df_bridge
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(
        "workspace.gold.bridge_evento_votacao"
    )
)
```

```python
%sql
OPTIMIZE workspace.gold.bridge_evento_votacao
ZORDER BY (
    id_evento,
    sk_votacao
)
```
