```python
# Imports
from pyspark.sql import functions as F
```

```python
# Leitura Silver
df = spark.table(
    "workspace.silver.partidos"
)

display(df)
```

```python
# Transformações Gold
df_gold = (

    df

    .select(
        F.col("id").alias("id_partido"),
        F.col("sigla"),
        F.col("nome"),
        F.col("uri")
    )

    .dropDuplicates()

    .withColumn(
        "dt_criacao_gold",
        F.current_timestamp()
    )

)
```

```python
# Validação
display(df_gold)
```

```python
# Escrita Gold
(
    df_gold.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(
        "workspace.gold.dim_partido"
    )
)
```

```python
%sql
-- Optimize
OPTIMIZE workspace.gold.dim_partido;
```
